```python
import os
import csv
import re

# Replace this with your actual local path to the image folder
GENERATED_DIR = "../generated"
CSV_FILE = "./generated_labels.csv"

# Mapping of benchmark name to category
CATEGORY_MAP = {
    "deutsch": "Hidden Subgroup",
    "iswap": "Logical Operation",
    "quantumwalks": "Quantum Walk",
    "grover": "Search and Optimization",
    "ipea": "Hidden Subgroup",
    "dnn": "Machine Learning",
    "teleportation": "Quantum Communication",
    "qaoa": "Search and Optimization",
    "toffoli": "Logical Operation",
    "linearsolver": "Linear Equation",
    "fredkin": "Logical Operation",
    "wstate": "Logical Operation",
    "basis_change": "Quantum Simulation",
    "qrng": "Quantum Arithmetic",
    "cat_state": "Logical Operation",
    "inverseqft": "Hidden Subgroup",
    "adder": "Quantum Arithmetic",
    "hs4": "Hidden Subgroup",
    "bell": "Logical Operation",
    "qft": "Hidden Subgroup",
    "variational": "Quantum Simulation",
    "vqe": "Linear Equation",
    "vqe_uccsd": "Linear Equation",
    "basis_trotter": "Quantum Simulation",
    "qec_sm": "Error Correction",
    "lpn": "Machine Learning",
    "qec_en": "Error Correction",
    "shor": "Hidden Subgroup",
    "pea": "Hidden Subgroup",
    "error_correctiond3": "Error Correction",
    "simons": "Hidden Subgroup",
    "hhl": "Linear Equation",
    "bb84": "Quantum Communication",
    "qpe": "Hidden Subgroup",
    "seca": "Error Correction",
    "sat": "Search and Optimization",
    "cc": "Search and Optimization",
    "multiply": "Quantum Arithmetic",
    "gcm": "Quantum Chemistry",
    "ising":"Quantum Simulation",
    "knn": "Search and Optimization",
    "bv": "Hidden Subgroup",
    "qf21": "Hidden Subgroup",
    "multiplier": "Quantum Arithmetic",
    "factor247": "Hidden Subgroup",
    "qec9xz": "Error Correction",
    "bigadder": "Quantum Arithmetic",
    "square_root": "Quantum Arithmetic",
    "qram": "Quantum Architecture",
    "bwt": "Quantum Walk",
    "cat": "Logical Operation",
    "ghz": "Logical Operation",
    "swap_test": "Machine Learning",
    "knn": "Search and Optimization",
    "qugan": "Machine Learning",
    "quantumtelecloning": "Quantum Cloning",
    "quantumvolume": "Quantum Volume",
    "qaoamax3sat": "QAOA"
}

image_labels = []

for filename in os.listdir(GENERATED_DIR):
    if filename.endswith(".png"):
        name_no_ext = filename.rsplit(".", 1)[0]
        if "_n" in name_no_ext:
            benchmark = name_no_ext.split("_n")[0].lower().replace("-", "").replace(" ", "").replace("_", "")
            category = CATEGORY_MAP.get(benchmark, "Unknown")
        else:
            benchmark = name_no_ext.lower().replace("-", "").replace(" ", "").replace("_", "")
            category = CATEGORY_MAP.get(benchmark, "Unknown")

        image_labels.append((filename, category))

# Save to CSV
with open(CSV_FILE, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["filename", "category"])
    writer.writerows(image_labels)

print(f"✅ Labels saved to {CSV_FILE} sdhjsdh")
```
